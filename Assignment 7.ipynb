# Question 1
##### Use explicit Euler and Predictor-Corrector methods to solve the following ODE, taking step sizes 0.5, 0.2 and 0.05. Plot the solutions for each step sizes in two different plots for different methods. (Take e = 2.71828)
##### dy/dx=y*ln{y}/x,   where y(2)=e

```python
def Func1(y,x):
    return(y*math.log(y)/x)
```

```python
XE1,YE1=Forward_Euler(Func1,2.71828,2,0.5,X_N=10)
XE2,YE2=Forward_Euler(Func1,2.71828,2,0.2,X_N=10)
XE3,YE3=Forward_Euler(Func1,2.71828,2,0.05,X_N=10)
plt.plot(XE1,YE1,"-",label="Step size=0.5")
plt.plot(XE2,YE2,"-",label="Step size=0.2")
plt.plot(XE3,YE3,"-",label="Step size=0.05")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Explicit Euler Method")
plt.legend(loc="upper left")
plt.grid()
plt.rcParams["figure.figsize"]=10,5
plt.show()
```

```python
XC1,YC1=Predictor_Corrector(Func1,2.71828,2,0.5,X_N=10)
XC2,YC2=Predictor_Corrector(Func1,2.71828,2,0.2,X_N=10)
XC3,YC3=Predictor_Corrector(Func1,2.71828,2,0.05,X_N=10)
plt.plot(XC1,YC1,"-",label="Step size=0.5")
plt.plot(XC2,YC2,"-",label="Step size=0.2")
plt.plot(XC3,YC3,"-",label="Step size=0.05")
plt.xlabel('x')
plt.ylabel('y')
plt.title("Predictor-Corrector Method")
plt.legend(loc="upper left")
plt.grid()
plt.rcParams["figure.figsize"]=10,5
plt.show()
```

# Question 2
##### Use RK4, for a step size of 0.05, to solve the equation,
##### d2y/dx2+dy/dx=1-x,      where y(0)=2,  y'(0)=1
##### Crosscheck your answer, by plotting, with the analytical solution y(x) = c1 +c2e^x  x^2/2 + 2x over the range x  [-5, 5] and y  [5, 5].

```python
def Func2(y, u, x):
    return(1-x-u)
```

```python
XRK4,YRK4=RK42nd(Func2,2,1,0,-5,5,0.05)
plt.plot(XRK4,YRK4,"g-.",label="RK4")
AE=[(1+math.exp(-x)-(x**2)/2+2*x) for x in XRK4]
plt.plot(XRK4,AE,"r:",label="Analytical")
plt.xlabel('(x)')
plt.ylabel('f(x)')
plt.legend(loc="upper right")
plt.show()
```
